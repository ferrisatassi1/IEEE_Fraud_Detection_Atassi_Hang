```python
# Standard Libraries
import math
from sklearn.utils import resample
import pickle
import warnings
# Scientific Computing and Data Manipulation
import numpy as np
import pandas as pd
import scipy
from scipy import stats
from scipy.spatial.distance import mahalanobis

# Machine Learning Models and Utilities
from sklearn.calibration import CalibratedClassifierCV, calibration_curve
from sklearn.decomposition import PCA
from sklearn.ensemble import (
    IsolationForest,
    RandomForestClassifier,
    StackingClassifier,
    VotingClassifier
)
from sklearn.experimental import enable_halving_search_cv
from sklearn.feature_selection import RFE, SelectFromModel, mutual_info_classif
from sklearn.linear_model import LogisticRegression, LassoCV
from sklearn.metrics import (
    ConfusionMatrixDisplay,
    accuracy_score,
    average_precision_score,
    balanced_accuracy_score,
    brier_score_loss,
    classification_report,
    confusion_matrix,
    f1_score,
    log_loss,
    matthews_corrcoef,
    precision_recall_curve,
    precision_score,
    recall_score,
    roc_auc_score,
    roc_curve,
    auc
)
from sklearn.model_selection import (
    GridSearchCV,
    HalvingRandomSearchCV,
    RandomizedSearchCV,
    RepeatedStratifiedKFold,
    cross_val_score,
    learning_curve,
    train_test_split
)
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import LabelEncoder, MinMaxScaler, StandardScaler, label_binarize
from sklearn.svm import OneClassSVM
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_text

# XGBoost
import xgboost as xgb
from xgboost import XGBClassifier

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns
import shap

# Imbalanced Learning
import imblearn
from imblearn.over_sampling import SMOTE

# Parallelization
from joblib import Parallel, delayed

# Google Colab Utilities
from google.colab import drive, files

# Warnings
warnings.filterwarnings('ignore')

# Google Drive Mount
drive.mount('/content/drive')
```

```python
# Paths to stored artifacts on Google Drive
features_path = '/content/drive/My Drive/IEEE_dataset/selected_features.pkl'
data_path = '/content/drive/My Drive/IEEE_dataset/preprocessed_data.pkl'
model_path = '/content/drive/My Drive/IEEE_dataset/voting_model_rf.pkl'
metrics_path = '/content/drive/My Drive/IEEE_dataset/final_metrics.pkl'

# Load Selected Features
with open(features_path, 'rb') as f:
    selected_features = pickle.load(f)

# Load Preprocessed Dataset
with open(data_path, 'rb') as f:
    data = pickle.load(f)
    X = data['X']
    y = data['y']

# Subset Data to Selected Features
X_selected = X[selected_features]

# Load Trained Voting Model
with open(model_path, 'rb') as f:
    voting_model = pickle.load(f)

# Optional: Load Final Metrics
with open(metrics_path, 'rb') as f:
    final_metrics = pickle.load(f)
    print("\nLoaded Final Metrics:")
    for metric, value in final_metrics.items():
        print(f"{metric.capitalize()}: {value:.4f}")

# Predict Using the Model
y_pred = voting_model.predict(X_selected)
y_prob = voting_model.predict_proba(X_selected)[:, 1] if hasattr(voting_model, 'predict_proba') else None

# Confusion Matrix and Error Analysis
tn, fp, fn, tp = confusion_matrix(y, y_pred).ravel()
print(f"\nConfusion Matrix:")
print(f"True Negatives: {tn}, False Positives (Type I Error): {fp}")
print(f"False Negatives (Type II Error): {fn}, True Positives: {tp}")

# Create Labels for Error Types
error_types = np.zeros_like(y)
error_types[(y == 1) & (y_pred == 0)] = 1  # Type II Error
error_types[(y == 0) & (y_pred == 1)] = 2  # Type I Error

# Train CART Decision Tree on Errors
cart_tree = DecisionTreeClassifier(random_state=42, max_depth=5)
cart_tree.fit(X_selected, error_types)

# Interpret Decision Tree
tree_rules = export_text(cart_tree, feature_names=list(X_selected.columns))
print("\nDecision Tree Rules for Error Analysis:")
print(tree_rules)

# Analyze Feature Importance for Errors
feature_importance = cart_tree.feature_importances_
important_features = pd.DataFrame({
    'Feature': X_selected.columns,
    'Importance': feature_importance
}).sort_values(by='Importance', ascending=False)

print("\nTop Features Correlated with Errors:")
print(important_features.head())

# Save Results for Future Analysis
important_features_path = '/content/drive/My Drive/IEEE_dataset/error_analysis_features.pkl'
important_features.to_pickle(important_features_path)
print(f"\nSaved important features to {important_features_path}")

# Generalization Analysis: Visualization of Metrics
metrics = pd.DataFrame(final_metrics, index=[0])
metrics.plot(kind='bar', figsize=(10, 6))
plt.title("Model Performance Metrics")
plt.ylabel("Metric Value")
plt.xticks(rotation=0)
plt.show()

# Distribution of Target Classes
plt.figure(figsize=(8, 6))
sns.countplot(x=y)
plt.title("Target Class Distribution")
plt.ylabel("Count")
plt.xlabel("Target Class")
plt.show()

# ROC Curve
if y_prob is not None:
    fpr, tpr, _ = roc_curve(y, y_prob)
    roc_auc = auc(fpr, tpr)

    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, label=f"ROC Curve (AUC = {roc_auc:.4f})")
    plt.plot([0, 1], [0, 1], 'k--', label="Random Guess")
    plt.title("ROC Curve")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.legend()
    plt.show()

# Correct vs Incorrect Predictions
correct_predictions = (y == y_pred).astype(int)
plt.figure(figsize=(8, 6))
sns.countplot(x=correct_predictions, palette="coolwarm")
plt.title("Correct vs Incorrect Predictions")
plt.ylabel("Count")
plt.xlabel("Prediction Correctness (1 = Correct, 0 = Incorrect)")
plt.show()

# Feature Importance Analysis: Top Features
plt.figure(figsize=(10, 8))
sns.barplot(data=important_features.head(10), x="Importance", y="Feature", palette="viridis")
plt.title("Top 10 Features Correlated with Errors")
plt.xlabel("Feature Importance")
plt.ylabel("Feature")
plt.show()
```

<p style="font-size:10px"> Using the outputs of our final model (the stacking model using Random Forest Feature Selection and repeated nested cross-validation), we then analyzed our errors to see which features contributed most significantly to them. We did this with a CART model, which had a very strong performance in determining which feature combinations resulted in errors, as seen in the ROC curve above. </p>

<p style="font-size:10px"> However, this strong performance may be deceptive, since it is based on nominal/categorical features which were numerically encoded in a way that may have made them falsely easy to split in the tree. Additionally, NaN values in feature columns were replaced with either the mean (for numerical features) or mode(for categorical features), which also may have created false calculability. For example, one of the feature splits on the preview above is for values of card1, which is the numerical code which differentiates different card brands (ex. CapitalOne Platinum vs. CapitalOne Quicksilver), but there is no reason why a higher or lower value for that code would be predictive of error other than simple coincidence. </p>

```python
# Subset Fraudulent Transactions
# Subset Fraudulent Transactions
fraud_data = X_selected[y == 1]  # Transactions labeled as fraud
fraud_caught = X_selected[(y == 1) & (y_pred == 1)]  # Fraud detected
fraud_missed = X_selected[(y == 1) & (y_pred == 0)]  # Fraud missed

print(f"\nFraudulent Transactions Analysis:")
print(f"Total Fraudulent Transactions: {len(fraud_data)}")
print(f"Fraudulent Transactions Caught (True Positives): {len(fraud_caught)}")
print(f"Fraudulent Transactions Missed (False Negatives): {len(fraud_missed)}")

# Compare feature distributions for selected features: C14, V307, Card_Transaction_Mean
selected_features_to_plot = ['C14', 'V307', 'Card_Transaction_Mean']
for feature in selected_features_to_plot:
    if feature in X_selected.columns:
        plt.figure(figsize=(8, 4))
        sns.kdeplot(fraud_caught[feature], label='Fraud Caught', color='green', fill=True, alpha=0.5)
        sns.kdeplot(fraud_missed[feature], label='Fraud Missed', color='red', fill=True, alpha=0.5)
        plt.title(f"Feature Distribution: {feature}")
        plt.xlabel(feature + " values")
        plt.ylabel("Probability Density")
        plt.legend()
        plt.show()
    else:
        print(f"Feature {feature} not found in the dataset.")


# Summary statistics for caught vs missed fraud
summary_stats = pd.DataFrame({
    'Caught_Mean': fraud_caught.mean(),
    'Caught_Std': fraud_caught.std(),
    'Missed_Mean': fraud_missed.mean(),
    'Missed_Std': fraud_missed.std()
})
summary_stats['Difference_Mean'] = summary_stats['Caught_Mean'] - summary_stats['Missed_Mean']
print("\nSummary Statistics for Caught vs Missed Fraud:")
print(summary_stats.sort_values(by='Difference_Mean', ascending=False).head(10))

# Train a Decision Tree on Fraud Data to Differentiate Caught vs Missed
fraud_labels = np.zeros_like(y[y == 1])  # Initialize labels
fraud_labels[y_pred[y == 1] == 1] = 1   # Label caught fraud as 1

cart_tree_fraud = DecisionTreeClassifier(random_state=42, max_depth=5)
cart_tree_fraud.fit(fraud_data, fraud_labels)

# Feature importance for fraud-specific analysis
fraud_feature_importance = pd.DataFrame({
    'Feature': X_selected.columns,
    'Importance': cart_tree_fraud.feature_importances_
}).sort_values(by='Importance', ascending=False)

print("\nTop Features Correlated with Fraud Detection Accuracy:")
print(fraud_feature_importance.head(10))

# Plot top 10 features for fraud detection
plt.figure(figsize=(10, 8))
sns.barplot(data=fraud_feature_importance.head(10), x="Importance", y="Feature", palette="mako")
plt.title("Top 10 Features Correlated with Fraud Detection Accuracy")
plt.xlabel("Feature Importance")
plt.ylabel("Feature")
plt.show()

# Analyzing transaction amounts for caught vs missed fraud
if 'TransactionAmount' in X_selected.columns:
    plt.figure(figsize=(8, 6))
    sns.boxplot(data=[fraud_caught['TransactionAmount'], fraud_missed['TransactionAmount']],
                notch=True, palette="coolwarm")
    plt.title("Transaction Amount: Fraud Caught vs Missed")
    plt.xticks([0, 1], ['Caught Fraud', 'Missed Fraud'])
    plt.ylabel("Transaction Amount")
    plt.show()
```

<p style="font-size:10px"> The above visualizations show different feature densities for correctly caught fraudulent transactions versus fraudulent transactions that were missed. In almost all cases, the feature density distribution was similar for the caught frauds versus the missed frauds, indicating that the differences are caused by interactions between features rather than any meaningful difference between the two classes (of caught frauds versus missed frauds). </p>

<p style="font-size:10px"> The CART model also generated a list of 10 features most predictive of whether a fraudulent transaction would be correctly classified versus incorrectly classified, as shown above. </p>

# 4) Results and Insights

Our model had fairly strong performance, with a ROC-AUC score of 0.94. Most of this was driven by the model's high precision, meaning that the vast majority (0.93) of transactions that were flagged as fraud were actually fraudulent. However, the model's recall was much weaker, at 0.43, meaning that it generated a significant number of false negatives. This was in line with the performance of the industry models examined in the literature review, which were all generally more effective in flagging only truly fraudulent transactions as fraudulent than in catching all fraudulent transactions. As discussed in the literature review, a preference for higher specificity and lower sensitivity might have business value, since it prevents customers from regularly having their transactions declined incorrectly.

In analyzing the most important features to fraud detection accuracy, we found that features from groups C1-C14 and D1-D15 were vastly disproportionately represented among the 10 most important features. Even though they made up less than 10% of the total features, they made up 50% of the 10 most important features. Features C1-C14 and D1-D15 often (though not always) encoded user characteristics or card usage patterns rather than transaction details, which seems to indicate that a more effective credit card fraud detection approach might emphasize detecting unusual user or card usage behavior instead of detecting unusual transctions. (It is worth noting that a similar approach focused on fraudulent users rather than fraudulent transactions won the IEEE-CIS Fraud Detection Kaggle competition that this dataset is drawn from.)

Finally, it is worth noting that the final stacked model incorporating XGBoost, Random Forest, and Logistic Regression together with the help of Random Forest Feature Selection using Repeated Nested Cross Validation only improved slightly (AUC of 0.94 versus 0.90) over simple XGBoost, even though it was much more computationally intensive and complex in terms of design. This provides evidence for the general principle that often model selection and design are not the important drivers of performance, and that what's important is instead feature selection, data hygiene, and other factors upstream of model design. With additional time, we would probably investigate if some of the methods we used for preprocessing might be changed to further improve importance.

# 5) Top 3 Features Correlated with Type 1 and 2 Errors

C14, V307, and Card_Transaction_Mean are the top 3 features related to causing errors in our model's prediction. However as explained earlier, this might be a result of correlations between features changing as the model progresses with preprocessing, feature engineering, and the overall imbalance in the dataset.

C14: This feature relates specifically to a count of specific occurrences seen across interactions. If the same phone number, card1-6 number, location, date, etc.., then the count feature will update accordingly to show their frequency. This is a Vesta Co. encoded feature, made to be unable to be clarified, as it regards privacy information. C14 specifically is an important count as it directly affects the model's judgement with the highest feature importance when error is made as a function of the feature space. To investigate this further, our team recommends corresponding with Vesta Co. to see what metrics they used to calculate it.

V307: This is another Vesta Co. encoded feature. With the V-features equating to calculated fields added to the dataset. Similar to C14, while working on this project we were unable to directly clarify what exactly V307 attempts to show other than the fact that it is a calculated value. This seems to be an important constructed feature that has a high impact on our dataset (noted in both the SHAP graph in file D and from error analysis in file F). To investigate this further, our team recommends corresponding with Vesta Co. to see what metrics they used to calculate it.

Card_Transaction_Mean is a feature that our group personally engineered. It is created through aggregating transaction amounts by Card_1 values, and obtaining the mean. One concern we have with this feature is its potential to cause overfitting as regardless of individual transactions. If a Card_1 value is the same across two diffferent transactions, same value will be seen in the two rows for that feature. This feature is important as it ties transaction amounts of transactions to card types, helping us indicate which cards tend to have higher/lower transaction amounts. To further improve this metric, we would add other Card_Transaction_Mean columns aggregated by the other card_# features.
